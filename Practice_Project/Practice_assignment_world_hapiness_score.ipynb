```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
z.shape
```

```python
df_new=df[(z<3).all(axis=1)]
print(df.shape)
print(df_new.shape)
```

```python
x=df_new.drop(columns=['Happiness Score'])
y=df_new['Happiness Score']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
scalar=StandardScaler()
X_scaled=scalar.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_scaled,y)
y_train.head()
```

```python
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
df_new.tail(7)
```

```python
print('Happiness Score is:',regression.predict(scalar.transform([[0.4,0.3,0.1,0.5,0.1,0.1,1.85]])))
```

```python
#saving the model on local file
import pickle
filename='finalized_model.pickle'
pickle.dump(regression,open(filename,'wb'))
#prediction using saved model
loaded_model=pickle.load(open(filename,'rb'))
a=loaded_model.predict(scalar.transform([[0.34,0.3,0.2,0.5,0.23,0.1,1.85]]))
a
```

```python
#adjusted R2 score
regression.score(x_train,y_train)
```
